## Abstract data type (ADT) 
### An abstract description of a data structure and the operations that can be performed on it.
- **Focuses on what** the data structure does, **not how** it does it.


- Specifies the **allowed operations** and their expected behavior, but **not the internal implementation details**.


- Like a **contract** defining what users can expect without revealing the inner workings.


- So a user only needs to know **what a data type can do**,\
but **not how it will be implemented**. Think of ADT as a black box which hides the inner structure and design of the data type.


- The user of data type does not need to know how that data type is implemented,\
 for example, we have been using Primitive values like `int`, `float`, `char` data types only with the knowledge that these data type can operate and be performed on without any idea of how they are implemented.


- Examples: List, Stack, Queue, Set, Map


## What are data structures?
### Data structures define how data is organized and stored in memory.


- They are mainly concerned with finding the best representation or organization of data In the memory that leads to efficient processing, manipulating and retrieving data.


- Defines **how** data is stored and accessed in memory.


- Uses specific algorithms to perform the operations specified by the ADT.


- Can be different implementations for the same ADT, each with its own performance characteristics.


- Examples: `array`, `vector`, `list`.

## Why data structures?
### Any computer software works with data!  


* How does google deal with millions of users?


* How does google maps find the shortest path between two locations?


* **Performance:** Choosing the right data structure directly impacts the efficiency of your program.


* **Memory Optimization:** Efficient data structures minimize memory usage, leading to better program performance and resource utilization.

## Algorithm analysis
### An **algorithm** is a clearly specified **set of simple instructions** to be followed to solve a problem.
- Once an algorithm is given for a problem and decided (somehow) to be correct, \
an important step is to determine **how much in the way of resources**, such as **time** or **space**,  
the algorithm will require. 


- An algorithm that solves a problem but **requires a year** is hardly of any use.  
Likewise, an algorithm that **requires thousands of gigabytes of main memory** is not (currently) useful on most machines.

- Algorithms must satisfy the following criteria:
    1. **Input** – Zero or more quantities are externally supplied. 
    
    2. **Output** – At least one quantity is produced.

    3. **Definiteness** – Each instruction is clear and unambiguous.

    4. **Finiteness** – If we trace out the instructions of an algorithm, then for all cases, the algorithm terminates after a finite number of steps.

    5. **Efficiency** – The algorithm must consume a reasonable amount of resources and takes a reasonable of time. It must be feasible to run the algorithm.

### A problem is defined in principle as a task that is amenable to being solved by a computer, e.g.
- Sorting a list of numbers.

- Searching for a particular key value in a collection of items.

- Finding the shortest distance between two cities in a map.
### We should differentiate between a problem (general description of a task) and its instances
- For example: 
    - **Problem** is to find the shortest distance between two cities.
    - **Problem instance** is to find the shortest distance between "El Korba" and "The 5th Settlement".

## Algorithm Complexity

### The term **complexity** refers to the amount of resources required by an algorithm to solve a _problem instance_.

- **Time complexity** refers to the amount of _time_ needed to solve a problem instance.

- **Space complexity** refers to the amount of _memory_ needed to solve a problem instance.

- Since an algorithm is a sequence of instructions written by a pencil on a piece of paper and cannot be executed directly to measure its performance,  
  we are faced with the following challenges:

  - How to measure an **algorithm performance**?

  - What are the **units of measurement**, is it seconds, or any other units?

- **Approach #1**

  - One can say Implement a program for this algorithm and run it for different instances of the problem under study.

    - Depends on the machine architecture, current technology, type of programming language used, the programmer skills, etc.

    - This approach is not practical!

- **Approach #2**

  - Select the most **fundamental operation** done by the algorithm,  
   then **count the number of times this operation takes place to solve a problem instance of size _n_**.

  - we use **the number of times the fundamental operation is executed** as a **measure of time complexity**.

  - It is **not measured in seconds or any other time units**.

    ```
    function add(arr, size)
    total = 0
    for i = 0 to size - 1
      total = total + arr[i]
    return total
    ```
  - People may choose **different fundamental operations** for the same algorithm, so you may get more than one time complexity function for the same algorithm!

  - Again, this approach depends on the **architecture** and the **technology** used, if for example we design a machine that executes * operation faster than + operation, our analysis will not be same!

  - Does it make any difference if someone tell you that the time complexity of an algorithm A is `T(n) = 3n+2` and somebody else insisted that it is `T(n) = 2n`?

  - Do you have to know the fundamental operation that the analysis is based on?

  - how about simplifying the time complexity function further by **ignoring all the coefficients**!?
    - So, if an algorithm has time complexity **T(n) = 4 $n^{2}$ + 100**, we simply say the time complexity is approximated to **T(n) ≈ $n^{2}$**
    


## Term Contribution
### Assume the actual time complexity of an algorithm is T(n) = 3 $n^{2}$ + 8 $n^{}$+ 10, what is the approximate time complexity of that algorithm?

- Since T(n) is getting bigger a.k.a **monotonically increasing** by increasing the problem size n,  
 we can study the contribution of each term **(3 $n^{2}$, 8 $n^{}$, and 10)**, on the increase of T(n).


````python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of values for n
n_values = np.arange(0, 101, 5)

# Define the function and calculate each term separately
def function(n):
    term1 = 3 * n ** 2
    term2 = 8 * n
    term3 = 10
    total = term1 + term2 + term3
    return term1/total, term2/total, term3/total

# Calculate ratios for each term and fill the lists
ratios_term1 = []
ratios_term2 = []
ratios_term3 = []
for n in n_values:
    ratio1, ratio2, ratio3 = function(n)
    ratios_term1.append(ratio1)
    ratios_term2.append(ratio2)
    ratios_term3.append(ratio3)

# Plot the ratios of each term over the total function value
plt.figure(figsize=(20, 6))

plt.plot(n_values, ratios_term1, label='3$n^2$ / (3$n^2$ + 8n + 10)', linewidth=3,color='red')
plt.plot(n_values, ratios_term2, label='8n / (3$n^2$ + 8n + 10)', linewidth=3,color='orange')
plt.plot(n_values, ratios_term3, label='10 / (3$n^2$ + 8n + 10)', linewidth=3,color='blue')

plt.xlabel('n')
plt.ylabel('Ratio')
plt.title('Ratio of Each Term over Total Function Value')
plt.legend()
plt.grid(True)
plt.xlim(0)
plt.ylim(0,1)

plt.show()
````

````python
import pandas as pd

# Define the range of values for n
n_values = range(0, 101, 5)


# Define the function and calculate each term separately
def function(n):
    term1 = 3 * n**2
    term2 = 8 * n
    term3 = 10
    total = term1 + term2 + term3
    return term1, term2, term3, total


# Create lists to store the data
data = {
    "n": [],
    "3n^2": [],
    "8n": [],
    "10": [],
    "Total": [],
    "3n^2%": [],
    "8n%": [],
    "10%": [],
}

# Calculate contributions for each term and fill the lists
for n in n_values:
    term1, term2, term3, total = function(n)
    data["n"].append(n)
    data["3n^2"].append(term1)
    data["8n"].append(term2)
    data["10"].append(term3)
    data["Total"].append(total)
    data["3n^2%"].append(round((term1 / total) * 100, 2))
    data["8n%"].append(round((term2 / total) * 100, 2))
    data["10%"].append(round((term3 / total) * 100, 2))

# Create DataFrame from the collected data
df = pd.DataFrame(data)

# Display the combined table
display(df)
````

- **Observation**:
  - As n 🡪 ∞ the term 3 $n^{2}$ dominates (i.e. approaches 100%) while the other terms decrease (i.e. approaches 0%)
- **Conclusion**:
  - We can **ignore the lower degree terms** from the complexity function as n 🡪 ∞  
    This leads to the first approximation of the previous complexity function to **T(n) ≈ 3 $n^{2}$**. 
### Now how about the coefficient 3? 
- If we ignore the **coefficient of the highest degree term**, it still **dominates the other two terms as $n_{}$ is getting bigger**.

````python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of values for n
n_values = np.arange(0, 101, 5)


# Define the function and calculate each term separately
def function(n):
    term1 = n**2
    term2 = 8 * n
    term3 = 10
    total = term1 + term2 + term3
    return term1 / total, term2 / total, term3 / total


# Calculate ratios for each term and fill the lists
ratios_term1 = []
ratios_term2 = []
ratios_term3 = []
for n in n_values:
    ratio1, ratio2, ratio3 = function(n)
    ratios_term1.append(ratio1)
    ratios_term2.append(ratio2)
    ratios_term3.append(ratio3)

# Plot the ratios of each term over the total function value
plt.figure(figsize=(20, 6))

plt.plot(
    n_values, ratios_term1, label="$n^2$ / (3$n^2$ + 8n + 10)", linewidth=3, color="red"
)
plt.plot(
    n_values, ratios_term2, label="8n / (3$n^2$ + 8n + 10)", linewidth=3, color="orange"
)
plt.plot(
    n_values, ratios_term3, label="10 / (3$n^2$ + 8n + 10)", linewidth=3, color="blue"
)

plt.xlabel("n")
plt.ylabel("Ratio")
plt.title("Ratio of Each Term over Total Function Value")
plt.legend()
plt.grid(True)
plt.xlim(0)
plt.ylim(0, 1)

plt.show()
````

- Observation:  
  Ignoring the coefficient of the highest degree term does not affect the contribution of that term on the growth of the complexity function T(n),     
  *i.e.* **it still dominates the other two terms as long as $n_{}$ 🡪 ∞**.

- Conclusion:  
  As **$n_{}$ 🡪 ∞** we can simply **drop the coefficient of the highest degree term**  
  since it is still dominating the other terms in the complexity function and therefore **T(n) ≈ $n^{2}$**. 

- So as the problem size $n_{}$ increases, we can approximate the complexity function,  
  but **what is the minimum problem size beyond which we can approximate**?
  
- *In the previous example*, we know that the exact time complexity is **T(n) = 3 $n^{2}$ + 8 $n^{}$ + 10** and the approximate one is **T(n) ≈ $n^{2}$**,  
  **what is the minimum problem size $n^{}$ that satisfies this approximation?**

### Draw Approximated T(n) over Exact T(n) as $n^{}$ increases

````python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of values for n
n_values = np.arange(0, 101, 1)

# Calculate the ratios
ratios1 = (n_values**2) / (3 * n_values**2 + 8 * n_values + 10)

# Plot the ratios
plt.figure(figsize=(15, 6))

plt.plot(
    n_values, ratios1, label=r"$\frac{n^2}{3n^2 + 8n + 10}$", linewidth=3, color="red"
)

plt.xlabel("n", fontsize=14)
plt.ylabel("Ratio", fontsize=14)
plt.title("Comparison of Ratios", fontsize=16)
plt.legend()
plt.xlim(0)
plt.ylim(0,.4)
plt.grid(True)
plt.show()
````

- **As you can see, when $n^{}$ is beyond a certain value, the ratio=0.33 starts to stabilize**.

- **We conclude from this experiment that the approximated T($n^{}$) is growing with the same rate as the actual T($n^{}$) when the problem size is around 25 or more**.

- **Generally speaking, mathematicians when they studied functions, they decided to group these functions according to their rate of growth**.

- **This field of science is known as asymptotic growth of functions**.

### We will focus on three important asymptotic notations:

- **Big-oh Notation: O(f($n^{}$))** - *worst case*

- **Theta Notation: θ(f($n^{}$))** - *average case*

- **Omega Notation: Ω(f($n^{}$))** - *best case*

### Big-oh Notation: O(f($n^{}$))

#### Definition

- **$g_{}$(N) = O($f_{}$(N)) if there are positive constants $c_{}$ and $n_{o}$ such that $g_{}$(N) ≤ $c_{}$ $f_{}$(N) when N ≥ $n_{o}$**.



````python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of values for N
N_values = np.arange(0, 20, 1)


# Define the functions g(N) and f(N)
def g(N):
    return 2 * N**2 + 5 * N + 10


def f(N):
    return N**2


# Calculate values of g(N) and c*f(N) for c = 3
g_values = g(N_values)
cf_values = 2.5 * f(N_values)

# Plot the functions
plt.figure(figsize=(20, 10))

plt.plot(N_values, g_values, label="g(N) = 2$N^2$ + 5N + 10", color="blue")
plt.plot(N_values, cf_values, label="cf(N) = 2.5$N^2$", color="red", linestyle="--")

plt.xlabel("N", fontsize=14)
plt.ylabel("Value", fontsize=14)
plt.title("Demonstration of g(N) = O(f(N))", fontsize=16)
plt.legend()
plt.xlim(0)
plt.ylim(0)
plt.grid(True)

plt.show()
````

### O( $f_{}$( $n_{}$)) = {  $g_{}$( $n_{}$) ∣ ∃ $c_{}$ > 0,∃  $n_{o}$  ​> 0,∀  $n_{}$ ≥  $n_{o}$ ​: 0 ≤  $g_{}$( $n_{}$) ≤ $c_{}$ $f_{}$( $n_{}$) } 
- **Although 1,000 $N^{}$ is larger than $N^{2}$ for small values of $N^{}$, $N^{2}$ grows at a faster rate, and**  
  **thus $N^{2}$ will eventually be the larger function. The turning point is $N^{}$ = 1,000 in this case**.

- The definition says that eventually **there is some point $n_{o}$ past which $c_{}$ · $f_{}$( $N^{}$) is always**  
  **at least as large as $g_{}$( $N^{}$)**, so that if constant factors are ignored, $f_{}$( $N^{}$) is at least as big as  
  $g_{}$( $N^{}$). In our case, we have $g_{}$( $N^{}$) = 1,000 $N^{}$, $f_{}$( $N^{}$) = $N^{2}$,  $n_{o}$ = 1,000, and $c_{}$ = 1. We could  
  also use  $n_{o}$ = 10 and $c_{}$ = 100. Thus, we can say that 1,000 $N^{}$ = O( $N^{2}$)

- We cannot say that 1,000 $N^{}$ = O( $N^{}$) **because $n_{o}$ = 0 and $c_{}$ = 1,001 for example**.

#### Typical growth rates

| Function                    | Name        |
|-----------------------------|-------------|
| $c^{}$                      | constant    |
| $log^{}$( $n^{}$)           | Logarithmic |
| $log^{}$( $n^{2}$)          | Log-squared |
| $n^{}$                      | Linear      |
| $n^{}$ $log^{}$( $n^{}$)    | Linear      |
| $n^{2}$                     | Quadratic   |
| $n^{3}$                     | Cubic       |
| $2^{n}$                     | Exponential |
| $!^{}$ $n^{}$               | Factorial   |

#### Rules
1. **If T(N) = O( $f_{}$(N)) and R(N) = O( $g_{}$(N))**, then  
    1. **T(N) + R(N) = O(  $f_{}$(N) + $g_{}$(N) )** intuitively and less formally it is **O(max( $f_{}$(N) , $g_{}$(N)))**. 
    
    2. **T(N) · R(N) = O(  $f_{}$(N) · $g_{}$(N) )**.
2. ***$log^{k}$( $n^{}$) = O( $n^{}$) for any constant k. This tells us that logarithms grow very slowly***  
  *because it is already known that N grows faster than any power of a log*.   
> we can always determine the relative growth rates of two functions $f_{}$(N) and  
  $g_{}$(N) by computing $lim_{N→∞}$ $f_{}$(N) / $g_{}$(N), using L’Hôpital’s rule if necessary.

### Ω and Θ Notations
#### Omega Ω definition
- **$g_{}$(N) = Ω($f_{}$(N)) if there are positive constants $c_{}$ and $n_{o}$ such that $g_{}$(N) ≥ $c_{}$ $f_{}$(N) when N ≥ $n_{o}$**.

- **$g_{}$(N) is Ω ($f_{}$(N)) if and only if $f_{}$(N) is O($g_{}$(N))**.

#### Theta Θ definition

- **T(N) = Θ(h(N)) if and only if T(N) is O(h(N)) and Ω(h(N))**.
